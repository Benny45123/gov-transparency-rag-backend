```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
#test conversation
```

```python
test_text = """
Q: Can you state your name for the record?
A: My name is Jeffrey Epstein.
Q: Were you present at the meeting in Paris on the 10th?
A: I do not recall that specific date, but I was in France that month.
Q: Who else was traveling with you on the private jet?
A: There were several associates, but I would need to check the flight logs for exact names.
"""
```

```python
splitter=RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=20, separators=["\n\n", "\n", " ", ""])
```

```python
chunks = splitter.split_text(test_text)
print(chunks)
for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1}:\n{chunk}\n")
```
